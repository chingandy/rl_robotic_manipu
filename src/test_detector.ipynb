```python
import gym
import random

env = gym.make("Reacher-v101")
state = env.reset()
data = []
for _ in range(50):
    action = random.choice(env.action_space)
    env.step(action)
    img = env.render(mode='rgb_array', width=256, height=256)
    data.append(img)
env.close()
```

### Test blob detector

```python
import numpy as np
import cv2
import matplotlib.pyplot as plt
%matplotlib inline
def blob_detector(im):

    """ Set up SimpleBlobDetector parameters """
    params = cv2.SimpleBlobDetector_Params()

    # Change thresholds
    params.minThreshold = 100
    params.maxThreshold = 200

    # Filter by Area
    params.filterByArea = True
    params.maxArea = 10

    # Filter by Circularity
    params.filterByCircularity = True
    params.minCircularity = 0.8

    # Filter by Convexity
    params.filterByConvexity = True
    params.minConvexity = 0.87

    # Filter by Inertia
    params.filterByInertia = True
    params.minInertiaRatio = 0.01

    # Set up the detector with default parameters
    detector = cv2.SimpleBlobDetector_create()
    

    # Detect blobs
    keypoints = detector.detect(im)
    im_with_keypoints = cv2.drawKeypoints(im, keypoints, np.array([]), (255, 255, 0), cv2.DRAW_MATCHES_FLAGS_DRAW_RICH_KEYPOINTS)
    plt.axis('off')
    plt.imshow(im_with_keypoints)
    plt.show()
    try:
        #if keypoints is not None:
         #   print("target pos: ", keypoints[0].pt)  
        return np.array(keypoints[0].pt)
    except IndexError:
        print("Detector failed.")
#         plt.axis('off')
#         plt.imshow(im)
#         plt.show()
        return np.zeros((1,2))

    
for img in data:
    print(blob_detector(img))
```
